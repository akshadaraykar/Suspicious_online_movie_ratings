# Fandango v/s all

```python
df = pd.merge(fandango,all_sites,on='FILM', how='inner')
df
```

```python
df.info()
```

```python
#normalization
df['RT_N'] = np.round(df['RottenTomatoes']/20,1)
df['RTU_N'] = np.round(df['RottenTomatoes_User']/20,1)
df
```

```python
df['Meta_N'] = np.round(df['Metacritic']/20,1)
df['MetaU_N'] = np.round(df['Metacritic_User']/2,1)
df['IMDB_N'] = np.round(df['IMDB']/2,1)
df
```

```python
df.columns
```

```python
df_N = df[['STARS','RATING','RT_N','RTU_N','Meta_N', 'MetaU_N', 'IMDB_N']]
df_N
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(data=df_N, x = 'RATING',fill = True, label = 'TRUE RATINGS')
sns.kdeplot(data=df_N, x = 'STARS', fill =True, label = 'STARS')
sns.kdeplot(data=df_N, x = 'RT_N',fill = True, label = 'RT_N')
sns.kdeplot(data=df_N, x = 'RTU_N', fill =True, label = 'RTU_N')
sns.kdeplot(data=df_N, x = 'Meta_N',fill = True, label = 'Meta_N')
sns.kdeplot(data=df_N, x = 'MetaU_N', fill =True, label = 'MetaU_N')
sns.kdeplot(data=df_N, x = 'IMDB_N',fill = True, label = 'IMDB_N')
plt.title('Fandango v/s All')
plt.legend(loc=(1.02,0.5))
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(data=df_N, x = 'STARS', fill =True, label = 'STARS')
sns.kdeplot(data=df_N, x = 'RT_N',fill = True, label = 'RT_N')
plt.legend(loc=(1.02,0.5))
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df_N)
```

```python
df_NF = df[['FILM','STARS','RATING','RT_N','RTU_N','Meta_N', 'MetaU_N', 'IMDB_N']]
df_NF['N_diff_RT_Stars'] = df['RT_N'] - df['STARS']
df_NF
```

```python
df10 = df_NF.nsmallest(10,'RT_N')
df10 
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(data=df10, x = 'RT_N',fill = True, label = 'RT_N')
sns.kdeplot(data=df10, x = 'STARS', fill =True, label = 'STARS')
sns.kdeplot(data=df10, x = 'RTU_N', fill =True, label = 'RTU_N')
sns.kdeplot(data=df10, x = 'Meta_N',fill = True, label = 'Meta_N')
sns.kdeplot(data=df10, x = 'IMDB_N',fill = True, label = 'IMDB_N')
sns.kdeplot(data=df10, x = 'RATING',fill = True, label = 'TRUE RATINGS')
plt.legend(loc=(1.02,0.5))
plt.show()
```
